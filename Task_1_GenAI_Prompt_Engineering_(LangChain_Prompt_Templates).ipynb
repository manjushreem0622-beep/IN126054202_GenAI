```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

## Task 1: Replace Hardcoded Prompts  

In this task, we replace static prompts with reusable PromptTemplate instead of using hardcoded strings.

```python
template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template.format(topic="Artificial Intelligence"))
```

## Task 2: Multi-Input Prompt System  

Here, we create a prompt template that accepts multiple inputs such as topic, audience, and tone.

```python
multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print(multi_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations Engine  

Different prompt styles are created to handle various use cases like teaching, interview, and storytelling.

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System  

ChatPromptTemplate allows role-based prompting using system and user messages.

```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

messages = chat_template.format_messages(role="teacher", topic="Neural Networks")

for msg in messages:
    print(msg.content)
```

## Task 5: Input Validation  

Input validation ensures that only valid audience and tone values are accepted.

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience → Setting default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → Setting default: formal")
        tone = "formal"

    return audience, tone

# Test Case 1 (Invalid)
audience, tone = validate_inputs("student", "cool")
print("Final Values:", audience, tone)

# Test Case 2 (Valid)
audience, tone = validate_inputs("beginner", "fun")
print("Final Values:", audience, tone)
```

## Task 6: Prompt Generator  

This function combines validation and prompt templates to generate dynamic prompts.

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} tone step by step",
        "interview": "Generate 3 interview questions on {topic} for {audience} in {tone} tone",
        "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone"
    }

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    return template.format(topic=topic, audience=audience, tone=tone)


print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability  

A single template is reused with multiple inputs to demonstrate scalability.

```python
reuse_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("Data Science", "expert", "casual"),
    ("Machine Learning", "beginner", "formal"),
    ("Deep Learning", "expert", "technical")
]

for t, a, tone in inputs:
    print(reuse_template.format(topic=t, audience=a, tone=tone))
```

## **Pipeline Flow (End-to-End Prompt Generation Process)**

User Input

   ↓

Validation Layer

   ↓

Prompt Template

   ↓

Dynamic Prompt Generation

   ↓

Output

```python
# Step 1: User Input
topic = "Artificial Intelligence"
audience = "student"
tone = "cool"
style = "teaching"

print("\nUser Input:")
print("Topic:", topic)
print("Audience:", audience)
print("Tone:", tone)
print("Style:", style)


# Step 2: Validation Function
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("\nInvalid audience → Setting default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → Setting default: formal")
        tone = "formal"

    return audience, tone


# Apply validation
audience, tone = validate_inputs(audience, tone)

print("\nAfter Validation:")
print("Audience:", audience)
print("Tone:", tone)


# Step 3: Prompt Template Selection
templates = {
    "teaching": "Explain {topic} for {audience} in a {tone} tone step by step",
    "interview": "Generate 3 interview questions on {topic} for {audience} in {tone} tone",
    "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone"
}

template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=templates[style]
)

print("\nSelected Prompt Template:")
print(templates[style])


# Step 4: Dynamic Prompt Generation
final_prompt = template.format(
    topic=topic,
    audience=audience,
    tone=tone
)

print("\nFinal Generated Prompt:")
print(final_prompt)
```

This example demonstrates the complete pipeline flow:

1. User provides input  
2. Inputs are validated and corrected  
3. Appropriate prompt template is selected  
4. Template is filled dynamically  
5. Final structured prompt is generated  

This shows how prompt engineering systems work in real-world AI applications.
