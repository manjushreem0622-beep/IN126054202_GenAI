## **Importing Required Libraries**

LangChain provides modular tools, and I am combining it with a transformer-based model.

```python
from transformers import pipeline
from langchain_core.prompts import PromptTemplate
```

## **Initializing the Language Model**

Here, I am using a lightweight open-source model (GPT-2) to simulate LLM behavior.

```python
llm = pipeline("text-generation", model="gpt2")
```

## **Step 1: Basic LLM Interaction**

Let me first observe how the model responds without any structured prompt.

This helps in understanding how the model behaves without any structured input.

```python
response = llm("Explain Artificial Intelligence in one simple sentence", max_length=50)
print(response[0]['generated_text'])
```

## **Step 2: Creating a Prompt Template**

Instead of writing prompts manually every time, I can design reusable templates.

```python
prompt = PromptTemplate.from_template(
    "Explain {topic} in a simple and beginner-friendly way"
)

formatted_prompt = prompt.format(topic="Machine Learning")
print(formatted_prompt)
```

## **Step 3: Building a Simple Pipeline**

Here I combine the prompt template with the language model.

This step connects the prompt and model, forming a basic workflow similar to LangChain pipelines.


```python
final_prompt = prompt.format(topic="Neural Networks")

response = llm(final_prompt, max_length=60)
print(response[0]['generated_text'])
```

Note: In modern LangChain, pipelines are created using a chaining mechanism where prompts and models are combined. Here, I simulate the same idea manually using a prompt template and a language model.

## **Step 4: Creating a Reusable Workflow**

Now I extend the same pipeline to generate real-world use cases.

```python
usecase_prompt = PromptTemplate.from_template(
    "Give 3 real-world applications of {topic}"
)

text = usecase_prompt.format(topic="Artificial Intelligence")

response = llm(text, max_length=80)
print(response[0]['generated_text'])
```

## **Step 5: Testing with Multiple Inputs**

To make the system dynamic, I test multiple topics.

```python
topics = ["Data Science", "Robotics", "Cyber Security"]

# Updated prompt (to avoid repetition and improve quality)
prompt = PromptTemplate.from_template(
    "Explain {topic} in a simple, clear and unique way with one real-world example"
)

for topic in topics:
    text = prompt.format(topic=topic)

    result = llm(
        text,
        max_length=80,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.7,
        pad_token_id=50256,
        truncation=True
    )

    print(f"\n🔹 Topic: {topic}")
    print(result[0]['generated_text'])
```

## **From these experiments, I observed:**

- The structure of a prompt significantly influences the quality of output  
- Reusable templates make workflows more efficient and scalable  
- Even simple models can generate meaningful responses when guided properly  
- Breaking the system into steps improves both understanding and control  
- Open-source models like GPT-2 sometimes produce repetitive or less accurate outputs compared to advanced LLMs

These observations highlight how LangChain focuses not just on models, but on designing structured AI workflows.
